```python
#Import Libs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Ignore Error Messages
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
#Import Database for EDA (After Upload the Database for the Google Colab)
df = pd.read_csv('/content/census.csv')
```

```python
#Database Visualization
print(df)
```

```python
#Show the Most Frequent Value in the Attribute
print(df['age'].mode())
print(df['workclass'].mode())
```

```python
#Example 01 of Graphics for Categorical Data
graph = px.treemap(df, path=['workclass', 'age'])
```

```python
#Show the Graphic
graph.show()
```

```python
#Example 02 of Graphics for Categorical Data
graph = px.treemap(df, path=['occupation', 'relationship', 'age'])
```

```python
#Show the Graphic
graph.show()
```

```python
#Example 03 of Graphics for Categorical Data
graph = px.parallel_categories(df, dimensions=['occupation', 'relationship'])
```

```python
#Show the Graphic
graph.show()
```

```python
#Example 04 of Graphics for Categorical Data
graph = px.parallel_categories(df, dimensions=['workclass', 'occupation', 'income'])
```

```python
#Show the Graphic
graph.show()
```

```python
#Example 05 of Graphics for Categorical Data
graph = px.parallel_categories(df, dimensions=['education', 'income'])
```

```python
#Show the Graphic
graph.show()
```
